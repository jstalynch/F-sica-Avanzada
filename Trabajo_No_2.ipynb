3. **Espira circular**

Una espira circular inmersa en el vacío y de radio $a$, ubicada en el plano $YZ$ y con centro en el eje de coordenadas. Si por la espira circula una corriente uniforme y constante $I_0$.

- Calcule  la  intensidad  de  campo  magnético  y  la  densidad  de  flujo  magnético  en  cualquier punto del eje axial de la espira. 
- Donde a es el último número de su cédula, si es $0$, tomar el penúltimo. 
- $I_0$ corresponde al promedio de su número de cédula. 

### Datos:
- Radio $a = 6$ m (último dígito de CC).
- Corriente $I_0$ : (promedio de cédula).

- Permeabilidad del vacío $\mu_0 = 4\pi \times 10^{-7}$ Tm/A.

<hr>

<h1>Solución analítica</h1>

### Ley de Biot-Savart
El campo magnético en un punto del eje axial $( x $) se obtiene integrando la contribución de cada elemento de corriente $( d\mathbf{l} $):

$$
\mathbf{B} = \frac{\mu_0 I_0}{4\pi} \oint \frac{d\mathbf{l} \times \mathbf{\hat{r}}}{r^2}
  (1) $$

donde $( \mathbf{\hat{r}} $) es el vector unitario desde $( d\mathbf{l} $) al punto de observación.

Simplificación para el eje \( x \)
Por simetría, solo la componente axial $( B_x $) no se cancela:

$$[
B_x = \frac{\mu_0 I_0 a^2}{2(a^2 + x^2)^{3/2}}
] (2)$$


Intensidad de campo magnético ($\mathbf{H}$)

En el vacío, la relación entre $\mathbf{B}$ y $\mathbf{H}$ es:

$$
\mathbf{H} = \frac{\mathbf{B}}{\mu_0}
$$

Por lo tanto, para la componente axial:

$$
H_x = \frac{B_x}{\mu_0} = \frac{I_0 a^2}{2(a^2 + x^2)^{3/2}} (3) \quad \text{(Intensidad de campo magnético)}
$$

*Valores máximos (en $x = 0$)*

$$[
B_{\text{max}} = \frac{\mu_0 I_0}{2a}, \quad H_{\text{max}} = \frac{I_0}{2a}
]$$

Relación física clave

$$[
B_x = \mu_0 H_x
]$$

<h1>Graficas</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats

# Parámetros
a = 6  # Radio (m)
cc = [1, 0, 0, 1, 4, 8, 0, 2, 8, 6]
I = stats.mean(cc)  # Corriente promedio en Amperios
mu0 = 4 * np.pi * 1e-7  # Permeabilidad del vacío (Tm/A)

# Eje x (desde -3a hasta 3a)
x = np.linspace(-3*a, 3*a, 500)

# Cálculo de H_x (Intensidad de campo magnético)
H_x = (I * a**2) / (2 * (a**2 + x**2)**(3/2))  # En A/m

# Cálculo de B_x (Densidad de flujo magnético)
B_x = mu0 * H_x  # En Tesla

# Crear figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Gráfica de H_x
ax1.plot(x, H_x, 'b-', linewidth=2, label=f'$H_x(x)$ (a = {a} m, $I_0$ = {I:.2f} A)')
ax1.set_title('Intensidad de campo magnético en el eje axial', fontsize=14)
ax1.set_ylabel('$H_x$ (A/m)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend(fontsize=12)

# Gráfica de B_x
ax2.plot(x, B_x, 'r-', linewidth=2, label=f'$B_x(x)$ ($\mu_0$ = {mu0:.2e} Tm/A)')
ax2.set_title('Densidad de flujo magnético en el eje axial', fontsize=14)
ax2.set_xlabel('Posición en el eje $x$ (m)', fontsize=12)
ax2.set_ylabel('$B_x$ (Tesla)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend(fontsize=12)

plt.tight_layout()
plt.show()

# Mostrar valores máximos
print(f"\nValores máximos:")
print(f"Intensidad de campo H_max = {max(H_x):.2e} A/m (en x = 0)")
print(f"Densidad de flujo B_max = {max(B_x):.2e} T (en x = 0)")
print(f"Corriente promedio I = {I:.2f} A")
```

## Análisis

- Reemplazando $x = 0$ en $(1)$ se obtiene  :  

  $$[
  B_{\text{max}} = \frac{\mu_0 I_0}{2a} = \frac{(4\pi \times 10^{-7})(4.5)}{2 \times 8} \approx 3.53 \times 10^{-7} \, \text{T}
  ]$$

- **Observaciones**  
  - En el centro de la espira ($x = 0$), las contribuciones de todos los elementos de corriente se suman de manera más constructiva
  - Tiene simetría con respecto al origen y a medida que se aleja del eje origen, disminuye la densidad.
- Reemplazando $x = 0$ en  $H_x$ se obtiene  :  

  $$[H_{\text{max}} = \frac{I_0}{2a} = \frac{3}{12} = 0.25\, \text{A/m}]$$

  - Fuera del centro, las componentes perpendiculares al eje $x$ se cancelan reduciendo la intensidad neta del campo, por eso en el centro se encuentra el máximo, esto se debe también por la simetría y se evidencia con la gráfica.

5. **Cilindro con densidad de corriente**

Un cilindro de radio $a$, inmerso en el vacío está orientado en el eje $Z$. Por este, circula una corriente cuya densidad está dada por $J$, y la permeabilidad del cilindro es de $3\mu_0$.

- Calcule la intensidad de campo magnético y la densidad de flujo magnético a lo largo del eje
radial del cilindro.
- Donde $a$ es el último número de su cédula, si es 0, tomar el penúltimo.

\begin{align*}
5.1 \quad J &= \frac{3r^3}{a^2} \\
5.2 \quad J &= \frac{2a^3}{2a^2 + r^2} \\
5.3 \quad J &= \frac{a^3}{5ar^3}
\end{align*}


<hr>

<h1>Solución analítica</h1>

Para este caso el Valor de $a$ es 6 por tanto las densidad $J$ es:

$$
\quad{J1}(r) = \frac{3r^3}{a^2}, \quad \text{con } a = 6     \Rightarrow \quad{J}(r) = \frac{3r^3}{6^2} = \frac{3r^3}{36} = \frac{r^3}{12}
$$

Usaremos la ley de Ampère en forma integral:

$$
\oint \vec{H} \cdot d\vec{l} = I_{\text{enc}}
$$

Debido a la simetría cilíndrica, tomamos una trayectoria circular de radio $r$ centrada en el eje del cilindro:

$$
\vec{H}(r) = H_{\phi}(r) \, \hat{\phi}
\quad \text{y} \quad
\oint \vec{H} \cdot d\vec{l} = H_{\phi}(r) \cdot 2\pi r
$$

Entonces:

$$
H_{\phi}(r) = \frac{I_{\text{enc}}(r)}{2\pi r}
$$

Ahora como ya tenemos despejado el campo magnetico, calculamos la corriente encerrada en el cilindro:

$$
\boxed{
\mathbf{I} = \int_S \vec{J} \cdot d\vec{s}
}
$$


$$
I_{\text{enc}}(r) = \int_0^r J(r') \cdot 2\pi r' \, dr'
$$

Sustituimos: $J(r') = \frac{r'^3}{12}$


$$
I_{\text{enc}}(r) = \int_0^r \frac{r'^3}{12} \cdot 2\pi r' \, dr' = \frac{2\pi}{12} \int_0^r r'^4 \, dr' = \frac{\pi}{6} \left[ \frac{r'^5}{5} \right]_0^r = \frac{\pi r^5}{30}
$$

Ahora suntituimos la corriente en la lay de Ampere

$$
H_{\phi}(r) = \frac{I_{\text{enc}}(r)}{2\pi r} = \frac{\pi r^5 / 30}{2\pi r} = \frac{r^4}{60}
$$

Entonces el campo magnético es:

$$
\vec{H}(r) = \frac{r^4}{60} \, \hat{\phi}
$$

Calculamos el flujo magnetico :

Sabemos que: $\vec{B} = \mu \vec{H}, \quad \mu = 3\mu_0$

así que:


$$
\vec{B}(r) = 3\mu_0 \cdot \frac{r^4}{60} \, \hat{\phi} = \frac{\mu_0 r^4}{20} \, \hat{\phi}
$$


###  Resultados para la densidad de corriente del ítem 5.1

- Intensidad del campo magnético:

$$
\boxed{
\vec{H}(r) = \frac{r^4}{60} \, \hat{\phi} \quad [\text{A/m}]
}
$$

- Densidad de flujo magnético:

$$
\boxed{
\vec{B}(r) = \frac{\mu_0 r^4}{20} \, \hat{\phi} \quad [\text{T}]
}
$$

---

Ahora calculamos intencidad del campo magnetico y la densidad del flujo para la densidad del item 5.2, como se trata del mismo procedimiento se resumen los pasos 

$$
J(r) = \frac{2a^3}{2a^2 + r^2}
\quad \text{con } a = 6 \Rightarrow
J(r) = \frac{2 \cdot 6^3}{2 \cdot 6^2 + r^2}
= \frac{432}{72 + r^2}
$$


Calculamos la corriente encerrada 

$$
I_{\text{enc}} = \int_0^r J(r') \cdot 2\pi r' \, dr' = \int_0^r \frac{432}{72 + r'^2} \cdot 2\pi r' \, dr' = 864\pi \int_0^r \frac{r'}{72 + r'^2} \, dr'
$$

Solución de la integral:


$$
\int \frac{r'}{72 + r'^2} \, dr' = \frac{1}{2} \ln(72 + r'^2)
$$

Evaluamos la integral:

$$
I_{\text{enc}} = 864\pi \cdot \left[ \frac{1}{2} \ln(72 + r'^2) \right]_0^r = 432\pi \left( \ln(72 + r^2) - \ln(72) \right) = 432\pi \ln\left( \frac{72 + r^2}{72} \right)
$$

Ahora calculamos el campo magnético 

$$
H(r) = \frac{I_{\text{enc}}}{2\pi r} = \frac{432\pi \ln\left( \frac{72 + r^2}{72} \right)}{2\pi r} = \frac{216 \ln\left( \frac{72 + r^2}{72} \right)}{r}
$$

$$
\boxed{
\vec{H}(r) = \frac{216 \ln\left( \frac{72 + r^2}{72} \right)}{r} \, \hat{\phi}
}
$$

calculamos la Densidad de flujo magnético:

$$
\vec{B}(r) = \mu \vec{H}(r) = 3\mu_0 \cdot \frac{216 \ln\left( \frac{72 + r^2}{72} \right)}{r} = \frac{648\mu_0 \ln\left( \frac{72 + r^2}{72} \right)}{r} \, \hat{\phi}
$$

$$
\boxed{
\vec{B}(r) = \frac{648\mu_0 \ln\left( \frac{72 + r^2}{72} \right)}{r} \, \hat{\phi}
}
$$

---

se toma la exprecion de la densidad de corriente del item 5.3 y se remplaza con 6 que es el ultimo dijito de la cedula y con ellos se realizan los mismo calculos de las nateriores densidades 

$$
J(r) = \frac{5r^3}{a^3}
\quad \text{con } a = 6 \Rightarrow
J(r) = \frac{5r^3}{6^3} = \frac{5r^3}{216}
$$

Se calcula la corriente encerrada 

$$
I_{\text{enc}} = \int_0^r J(r') \cdot 2\pi r' \, dr'
= \int_0^r \frac{5r'^3}{216} \cdot 2\pi r' \, dr'
= \frac{10\pi}{216} \int_0^r r'^4 \, dr'
$$

Resolvemos la integral:

$$
= \frac{10\pi}{216} \left[ \frac{r'^5}{5} \right]_0^r = \frac{10\pi}{216} \cdot \frac{r^5}{5}
= \frac{2\pi r^5}{216}
$$

Ahora pasamos a calcular el campo magnético:

$$
H(r) = \frac{I_{\text{enc}}}{2\pi r} = \frac{2\pi r^5 / 216}{2\pi r} = \frac{r^4}{216}
$$

$$
\boxed{
\vec{H}(r) = \frac{r^4}{216} \, \hat{\phi}
}
$$

Por ultimo calculamos la densidad de flujo magnético:

$$
\vec{B}(r) = \mu \vec{H}(r) = 3\mu_0 \cdot \frac{r^4}{216} = \frac{\mu_0 r^4}{72}
$$

$$
\boxed{
\vec{B}(r) = \frac{\mu_0 r^4}{72} \, \hat{\phi}
}
$$

<h1>Graficas</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const

# Parámetros físicos
mu_0 = const.mu_0  # permeabilidad del vacío
a = 6.0             # radio del cilindro

# Dominio radial (evitamos r=0 por división)
r = np.linspace(0.01, a, 500)

# Ítem 5.1
H_5_1 = r**4 / 60
B_5_1 = mu_0 * r**4 / 20

# Ítem 5.2
# J(r) = 432 / (72 + r^2)
# I_enc derivado -> H(r) = [216 * ln((72 + r^2)/72)] / r
H_5_2 = 216 * np.log((72 + r**2) / 72) / r
B_5_2 = 648 * mu_0 * np.log((72 + r**2) / 72) / r

# Ítem 5.3
H_5_3 = r**4 / 216
B_5_3 = mu_0 * r**4 / 72

# Crear figura
plt.figure(figsize=(12, 8))

# Subplot 1: Campo magnético H(r)
plt.subplot(2, 1, 1)
plt.plot(r, H_5_1, label='H(r) - Ítem 5.1')
plt.plot(r, H_5_2, label='H(r) - Ítem 5.2')
plt.plot(r, H_5_3, label='H(r) - Ítem 5.3')
plt.title('Comparación de la Intensidad del Campo Magnético H(r)')
plt.xlabel('r [m]')
plt.ylabel('H(r) [A/m]')
plt.grid(True)
plt.legend()

# Subplot 2: Densidad de flujo magnético B(r)
plt.subplot(2, 1, 2)
plt.plot(r, B_5_1, label='B(r) - Ítem 5.1')
plt.plot(r, B_5_2, label='B(r) - Ítem 5.2')
plt.plot(r, B_5_3, label='B(r) - Ítem 5.3')
plt.title('Comparación de la Densidad de Flujo Magnético B(r)')
plt.xlabel('r [m]')
plt.ylabel('B(r) [T]')
plt.grid(True)
plt.legend()

# Ajuste de diseño y visualización
plt.tight_layout()
plt.show()
```

<h1>Análisis</h1>

Las gráficas muestran la comparación entre la intensidad del campo magnético $H(r)$ y la densidad de flujo magnético $B(r)$ para los tres casos estudiados. En el caso del ítem 5.1, $H(r)$ crece con la cuarta potencia de la distancia radial $r$, mostrando un crecimiento lento en las cercanías del eje y acelerado al acercarse al radio máximo $a = 6$. Este comportamiento genera una curva pronunciada que alcanza valores altos cerca del límite del cilindro. El ítem 5.3 presenta un crecimiento similar, también polinómico en $r^4$, pero con un coeficiente menor, lo que produce una curva más suave y de menor magnitud en toda la región.

Por otro lado, el ítem 5.2 exhibe un comportamiento notablemente diferente: su intensidad de campo está modelada por una función logarítmica dividida por $r$. Esto implica que cerca del eje, la función $H(r)$ inicia con valores relativamente altos y crece rápidamente, pero a medida que $r$ aumenta, la curva se aplana, mostrando una tendencia a la saturación o estabilización. Este comportamiento puede reflejar efectos más realistas en medios con propiedades no lineales o límites físicos en la respuesta electromagnética.

En cuanto a la densidad de flujo magnético $B(r)$, sus curvas replican la forma de $H(r)$, solo que multiplicadas por la permeabilidad del medio, que en este caso es $3\mu_0$. Así, mantienen el mismo patrón de crecimiento y saturación, pero con magnitudes típicas para campos magnéticos bajos, como los que se encuentran en aplicaciones biomédicas o experimentos de laboratorio.

En resumen, los ítems 5.1 y 5.3 describen modelos ideales con distribuciones polinómicas de corriente que producen campos crecientes y sin límite en el rango analizado, mientras que el ítem 5.2 propone un modelo más complejo que incorpora una saturación implícita vía la dependencia logarítmica, representando condiciones más realistas para ciertos materiales o tejidos. Esta diferencia hace que el ítem 5.2 sea potencialmente más útil para simulaciones o estudios donde se consideren efectos físicos limitantes, mientras que los ítems 5.1 y 5.3 son útiles para análisis matemáticos ideales o aproximaciones iniciales.

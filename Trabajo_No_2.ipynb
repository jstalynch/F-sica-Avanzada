5. **Cilindro con densidad de corriente**

Un cilindro de radio $a$, inmerso en el vacío está orientado en el eje $Z$. Por este, circula una corriente cuya densidad está dada por $J$, y la permeabilidad del cilindro es de $3\mu_0$.

- Calcule la intensidad de campo magnético y la densidad de flujo magnético a lo largo del eje
radial del cilindro.
- Donde $a$ es el último número de su cédula, si es 0, tomar el penúltimo.

\begin{align*}
5.1 \quad J &= \frac{3r^3}{a^2} \\
5.2 \quad J &= \frac{2a^3}{2a^2 + r^2} \\
5.3 \quad J &= \frac{a^3}{5ar^3}
\end{align*}


<hr>

<h1>Solución analítica</h1>

Para este caso el Valor de $a$ es 6 por tanto las densidad $J$ es:

$$
\quad{J1}(r) = \frac{3r^3}{a^2}, \quad \text{con } a = 6     \Rightarrow \quad{J}(r) = \frac{3r^3}{6^2} = \frac{3r^3}{36} = \frac{r^3}{12}
$$

Usaremos la ley de Ampère en forma integral:

$$
\oint \vec{H} \cdot d\vec{l} = I_{\text{enc}}
$$

Debido a la simetría cilíndrica, tomamos una trayectoria circular de radio $r$ centrada en el eje del cilindro:

$$
\vec{H}(r) = H_{\phi}(r) \, \hat{\phi}
\quad \text{y} \quad
\oint \vec{H} \cdot d\vec{l} = H_{\phi}(r) \cdot 2\pi r
$$

Entonces:

$$
H_{\phi}(r) = \frac{I_{\text{enc}}(r)}{2\pi r}
$$

Ahora como ya tenemos despejado el campo magnetico, calculamos la corriente encerrada en el cilindro:

$$
\boxed{
\mathbf{I} = \int_S \vec{J} \cdot d\vec{s}
}
$$


$$
I_{\text{enc}}(r) = \int_0^r J(r') \cdot 2\pi r' \, dr'
$$

Sustituimos: $J(r') = \frac{r'^3}{12}$


$$
I_{\text{enc}}(r) = \int_0^r \frac{r'^3}{12} \cdot 2\pi r' \, dr' = \frac{2\pi}{12} \int_0^r r'^4 \, dr' = \frac{\pi}{6} \left[ \frac{r'^5}{5} \right]_0^r = \frac{\pi r^5}{30}
$$

Ahora suntituimos la corriente en la lay de Ampere

$$
H_{\phi}(r) = \frac{I_{\text{enc}}(r)}{2\pi r} = \frac{\pi r^5 / 30}{2\pi r} = \frac{r^4}{60}
$$

Entonces el campo magnético es:

$$
\vec{H}(r) = \frac{r^4}{60} \, \hat{\phi}
$$

Calculamos el flujo magnetico :

Sabemos que: $\vec{B} = \mu \vec{H}, \quad \mu = 3\mu_0$

así que:


$$
\vec{B}(r) = 3\mu_0 \cdot \frac{r^4}{60} \, \hat{\phi} = \frac{\mu_0 r^4}{20} \, \hat{\phi}
$$


###  Resultados para la densidad de corriente del ítem 5.1

- **Intensidad del campo magnético:**

$$
\boxed{
\vec{H}(r) = \frac{r^4}{60} \, \hat{\phi} \quad [\text{A/m}]
}
$$

- **Densidad de flujo magnético:**

$$
\boxed{
\vec{B}(r) = \frac{\mu_0 r^4}{20} \, \hat{\phi} \quad [\text{T}]
}
$$

---

Ahora calculamos intencidad del campo magnetico y la densidad del flujo para la densidad del item 5.2, como se trata del mismo procedimiento se resumen los pasos 

$$
J(r) = \frac{2a^3}{2a^2 + r^2}
\quad \text{con } a = 6 \Rightarrow
J(r) = \frac{2 \cdot 6^3}{2 \cdot 6^2 + r^2}
= \frac{432}{72 + r^2}
$$


Calculamos la corriente encerrada 

$$
I_{\text{enc}} = \int_0^r J(r') \cdot 2\pi r' \, dr' = \int_0^r \frac{432}{72 + r'^2} \cdot 2\pi r' \, dr' = 864\pi \int_0^r \frac{r'}{72 + r'^2} \, dr'
$$

Solución de la integral:


$$
\int \frac{r'}{72 + r'^2} \, dr' = \frac{1}{2} \ln(72 + r'^2)
$$

Evaluamos la integral:

$$
I_{\text{enc}} = 864\pi \cdot \left[ \frac{1}{2} \ln(72 + r'^2) \right]_0^r = 432\pi \left( \ln(72 + r^2) - \ln(72) \right) = 432\pi \ln\left( \frac{72 + r^2}{72} \right)
$$

Ahora calculamos el campo magnético 

$$
H(r) = \frac{I_{\text{enc}}}{2\pi r} = \frac{432\pi \ln\left( \frac{72 + r^2}{72} \right)}{2\pi r} = \frac{216 \ln\left( \frac{72 + r^2}{72} \right)}{r}
$$

$$
\boxed{
\vec{H}(r) = \frac{216 \ln\left( \frac{72 + r^2}{72} \right)}{r} \, \hat{\phi}
}
$$

calculamos la Densidad de flujo magnético:

$$
\vec{B}(r) = \mu \vec{H}(r) = 3\mu_0 \cdot \frac{216 \ln\left( \frac{72 + r^2}{72} \right)}{r} = \frac{648\mu_0 \ln\left( \frac{72 + r^2}{72} \right)}{r} \, \hat{\phi}
$$

$$
\boxed{
\vec{B}(r) = \frac{648\mu_0 \ln\left( \frac{72 + r^2}{72} \right)}{r} \, \hat{\phi}
}
$$

---

se toma la exprecion de la densidad de corriente del item 5.3 y se remplaza con 6 que es el ultimo dijito de la cedula y con ellos se realizan los mismo calculos de las nateriores densidades 

$$
J(r) = \frac{5r^3}{a^3}
\quad \text{con } a = 6 \Rightarrow
J(r) = \frac{5r^3}{6^3} = \frac{5r^3}{216}
$$

Se calcula la corriente encerrada 

$$
I_{\text{enc}} = \int_0^r J(r') \cdot 2\pi r' \, dr'
= \int_0^r \frac{5r'^3}{216} \cdot 2\pi r' \, dr'
= \frac{10\pi}{216} \int_0^r r'^4 \, dr'
$$

Resolvemos la integral:

$$
= \frac{10\pi}{216} \left[ \frac{r'^5}{5} \right]_0^r = \frac{10\pi}{216} \cdot \frac{r^5}{5}
= \frac{2\pi r^5}{216}
$$

Ahora pasamos a calcular el campo magnético:

$$
H(r) = \frac{I_{\text{enc}}}{2\pi r} = \frac{2\pi r^5 / 216}{2\pi r} = \frac{r^4}{216}
$$

$$
\boxed{
\vec{H}(r) = \frac{r^4}{216} \, \hat{\phi}
}
$$

Por ultimo calculamos la densidad de flujo magnético:

$$
\vec{B}(r) = \mu \vec{H}(r) = 3\mu_0 \cdot \frac{r^4}{216} = \frac{\mu_0 r^4}{72}
$$

$$
\boxed{
\vec{B}(r) = \frac{\mu_0 r^4}{72} \, \hat{\phi}
}
$$

<h1>Graficas</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const

# Parámetros físicos
mu_0 = const.mu_0  # permeabilidad del vacío
a = 6.0             # radio del cilindro

# Dominio radial (evitamos r=0 por división)
r = np.linspace(0.01, a, 500)

# Ítem 5.1
H_5_1 = r**4 / 60
B_5_1 = mu_0 * r**4 / 20

# Ítem 5.2
# J(r) = 432 / (72 + r^2)
# I_enc derivado -> H(r) = [216 * ln((72 + r^2)/72)] / r
H_5_2 = 216 * np.log((72 + r**2) / 72) / r
B_5_2 = 648 * mu_0 * np.log((72 + r**2) / 72) / r

# Ítem 5.3
H_5_3 = r**4 / 216
B_5_3 = mu_0 * r**4 / 72

# Crear figura
plt.figure(figsize=(12, 8))

# Subplot 1: Campo magnético H(r)
plt.subplot(2, 1, 1)
plt.plot(r, H_5_1, label='H(r) - Ítem 5.1')
plt.plot(r, H_5_2, label='H(r) - Ítem 5.2')
plt.plot(r, H_5_3, label='H(r) - Ítem 5.3')
plt.title('Comparación de la Intensidad del Campo Magnético H(r)')
plt.xlabel('r [m]')
plt.ylabel('H(r) [A/m]')
plt.grid(True)
plt.legend()

# Subplot 2: Densidad de flujo magnético B(r)
plt.subplot(2, 1, 2)
plt.plot(r, B_5_1, label='B(r) - Ítem 5.1')
plt.plot(r, B_5_2, label='B(r) - Ítem 5.2')
plt.plot(r, B_5_3, label='B(r) - Ítem 5.3')
plt.title('Comparación de la Densidad de Flujo Magnético B(r)')
plt.xlabel('r [m]')
plt.ylabel('B(r) [T]')
plt.grid(True)
plt.legend()

# Ajuste de diseño y visualización
plt.tight_layout()
plt.show()
```

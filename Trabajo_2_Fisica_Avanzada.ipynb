#Campo magnético en una espira circular

## Problema 1:

Una espira circular de radio $a$, ubicada en el plano $YZ$ y centrada en el origen, transporta una corriente constante $I_0$.
Usando CC:1001480286

### Datos:
- Radio $a = 6$ m (último dígito de CC).
- Corriente $I_0$ : (promedio de cédula).

- Permeabilidad del vacío $\mu_0 = 4\pi \times 10^{-7}$ Tm/A.

### Objetivo:
Calcular la densidad de flujo magnético $\mathbf{B}$ en cualquier punto del eje $x$ y también la intensidad de campo $\mathbf{H}$


## Solución analítica

### Ley de Biot-Savart
El campo magnético en un punto del eje axial $( x $) se obtiene integrando la contribución de cada elemento de corriente $( d\mathbf{l} $):

$$
\mathbf{B} = \frac{\mu_0 I_0}{4\pi} \oint \frac{d\mathbf{l} \times \mathbf{\hat{r}}}{r^2}
  (1) $$

donde $( \mathbf{\hat{r}} $) es el vector unitario desde $( d\mathbf{l} $) al punto de observación.

### Simplificación para el eje \( x \)
Por simetría, solo la componente axial $( B_x $) no se cancela:

$[
B_x = \frac{\mu_0 I_0 a^2}{2(a^2 + x^2)^{3/2}}
] (2)$


###Intensidad de campo magnético ($\mathbf{H}$)

En el vacío, la relación entre $\mathbf{B}$ y $\mathbf{H}$ es:

$
\mathbf{H} = \frac{\mathbf{B}}{\mu_0}
$

Por lo tanto, para la componente axial:

$
H_x = \frac{B_x}{\mu_0} = \frac{I_0 a^2}{2(a^2 + x^2)^{3/2}} (3) \quad \text{(Intensidad de campo magnético)}
$

###Valores máximos (en $x = 0$)

$[
B_{\text{max}} = \frac{\mu_0 I_0}{2a}, \quad H_{\text{max}} = \frac{I_0}{2a}
$]

### Relación física clave

$[
B_x = \mu_0 H_x
$]


```python
import numpy as np
import mlib.pyplot as plt
import statisatplottics as s

# Parámetros
a = 6  # Radio (m)
cc = [1, 0, 0, 1, 4, 8, 0, 2, 8, 6]
I = s.mean(cc)  # Corriente promedio en Amperios
mu0 = 4 * np.pi * 1e-7  # Permeabilidad del vacío (Tm/A)

# Eje x (desde -3a hasta 3a)
x = np.linspace(-3*a, 3*a, 500)

# Cálculo de H_x (Intensidad de campo magnético)
H_x = (I * a**2) / (2 * (a**2 + x**2)**(3/2))  # En A/m

# Cálculo de B_x (Densidad de flujo magnético)
B_x = mu0 * H_x  # En Tesla

# Crear figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Gráfica de H_x
ax1.plot(x, H_x, 'b-', linewidth=2, label=f'$H_x(x)$ (a = {a} m, $I_0$ = {I:.2f} A)')
ax1.set_title('Intensidad de campo magnético en el eje axial', fontsize=14)
ax1.set_ylabel('$H_x$ (A/m)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend(fontsize=12)

# Gráfica de B_x
ax2.plot(x, B_x, 'r-', linewidth=2, label=f'$B_x(x)$ ($\mu_0$ = {mu0:.2e} Tm/A)')
ax2.set_title('Densidad de flujo magnético en el eje axial', fontsize=14)
ax2.set_xlabel('Posición en el eje $x$ (m)', fontsize=12)
ax2.set_ylabel('$B_x$ (Tesla)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend(fontsize=12)

plt.tight_layout()
plt.show()

# Mostrar valores máximos
print(f"\nValores máximos:")
print(f"Intensidad de campo H_max = {max(H_x):.2e} A/m (en x = 0)")
print(f"Densidad de flujo B_max = {max(B_x):.2e} T (en x = 0)")
print(f"Corriente promedio I = {I:.2f} A")
```

## Resultados y Análisis

- Reemplazando $x = 0$ en $(1)$ se obtiene  :  

  $[
  B_{\text{max}} = \frac{\mu_0 I_0}{2a} = \frac{(4\pi \times 10^{-7})(4.5)}{2 \times 8} \approx 3.53 \times 10^{-7} \, \text{T}
  $]

- **Observaciones**  
  - En el centro de la espira ($x = 0$), las contribuciones de todos los elementos de corriente se suman de manera más constructiva
  - Tiene simetría con respecto al origen y a medida que se aleja del eje origen, disminuye la densidad.
- Reemplazando $x = 0$ en  $H_x$ se obtiene  :  

  $[
H_{\text{max}} = \frac{I_0}{2a} = \frac{3}{12} = 0.25\, \text{A/m}
$]

- **Observaciones**  
  - Fuera del centro, las componentes perpendiculares al eje $x$ se cancelan reduciendo la intensidad neta del campo, por eso en el centro se encuentra el máximo, esto se debe también por la simetría y se evidencia con la gráfica.
